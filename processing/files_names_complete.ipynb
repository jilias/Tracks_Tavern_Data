```python
# import dependencis
import pandas as pd
import re
import datetime
```

```python
# read files path csv file

path_df = pd.read_csv('../Resources/sales_data_path.csv')
path_df.head()
```

```python
# Parse file name from path
path_df['file_name'] = path_df['Path'].str.extract(r"(itemPerformanceReport\d+.csv)")
path_df.head()
```

```python
# Parse Date as String from File name
path_df['date_str'] = path_df['file_name'].str.extract(r'(\d+)')
path_df
```

```python
# Convert Date string into datetime format
rows =[]
r = len(path_df['date_str'])
for i in range(r):
    rows.append(pd.to_datetime(path_df['date_str'].values[i], format='%m%d%Y'))
path_df['date'] = rows

path_df.dtypes
```

```python
# Drop date string
path_df.drop(['date_str'],axis=1, inplace=True)
path_df.head()
```

```python
# Drop Path column
path_df.drop(['Path'],axis=1, inplace=True)
path_df.head()
```

```python
# Export to csv file
path_df.to_csv('../Resources/file_names.csv', index=False)
```
