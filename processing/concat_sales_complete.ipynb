```python
# import dependencies
import pandas as pd
import datetime
```

```python
# Read csv files with files names
file_names = pd.read_csv('../Resources/file_names.csv')
file_names.head()
```

```python
# Create CSV Files with Dates
r = len(file_names['file_name'])
for file in range(r):
    path = 'Resources/weekly_sales/'
    csv_name = file_names['file_name'].values[file]
    columns = ['Item', "Item Code", "Quantity", "Unit Price", "Total Sales Amount"]
    sales_wk = pd.read_csv(f"{path}{csv_name}",header=1, usecols=columns)
    sales_wk['date'] = file_names['date'].values[file]
    sales_wk.to_csv(f'Resources/weekly_sales_dates/{csv_name}', index=False)
```

```python
# Create Dataframes with each weekly sales csv file
frames = []
for df in range(r):
    path2 = '../Resources/weekly_sales_dates/'
    csv_name2 = file_names['file_name'].values[df]
    frames.append(pd.read_csv(f"{path2}{csv_name2}"))

frames[100].head()
```

```python
# Concatonate the DF
weekly_sales_df = pd.concat(frames)
weekly_sales_df
```

```python
# Export to csv file
weekly_sales_df.to_csv('../Resources/weekly_sales.csv', index=False)
```
